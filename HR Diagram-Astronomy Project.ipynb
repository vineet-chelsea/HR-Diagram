<h5>The Mass Lifetime Relation for Main sequence Stars</h5>
<p>A star's lifetime on the main sequence is how long it takes to use up the hydrogen in its
core. The luminosity (L) of a star is a measure of how rapidly it is using up its hydrogen.
The mass (M) of a star is a measure of how much fuel it has. The time it takes to use up
the fuel is proportional to its amount of fuel (M) divided by the rate of fuel consumption
(L)</p>
<p>Also,Luminosity is related to 3.5 raised to power of Mass,therefore the lifetime ultimately depends upon M<sup>-2.5</sup> factor of a star</p>

<h2>Generating a colour Magnitude Diagram</h2>

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from astropy.io import fits
```

```python
#importing all the given files
# 3 images from b and v filter have been given respectively.Also,Master Bias and Flat Norm files have also been provided
b1 = fits.open('M29_B_ExpTime10s.fit') #image through blue filter at an exposure time of 10s
b2 = fits.open('M29_B_ExpTime30s.fit') #image through blue filter at an exposure time of 30s
b3 = fits.open('M29_B_ExpTime60s.fit') #image through blue filter at an exposure time of 60s
v1 = fits.open('M29_V_ExpTime10s.fit') #image through visual filter at an exposure time of 10s
v2 = fits.open('M29_V_ExpTime30s.fit') #image through visual filter at an exposure time of 10s
v3 = fits.open('M29_V_ExpTime60s.fit') #image through visual filter at an exposure time of 10s
mb = fits.open('MasterBias.fit') #master bias 
nfb = fits.open('NormFlatB.fit') # for normalisation B filter
nfv = fits.open('NormFlatV.fit') # for normalisation V filter
b1.info() #information regarding the file and dimesnsions
```

```python
b1[0].header #detailed information with date-time,exposurem,filter,date,Instrument,etc.
```

```python
#viewing the image through B filter
plt.style.use('dark_background')
plt.subplot(2,2,1)
plt.imshow(b1[0].data,cmap ='gray')
plt.subplot(2,2,2)
plt.imshow(b2[0].data,cmap = 'graby')
plt.subplot(2,2,3)
plt.imshow(b3[0].data,cmap = 'gray')
```

```python
#viewing the image through the V Filter
plt.style.use('dark_background')
plt.subplot(2,2,1)
plt.imshow(v1[0].data,cmap ='gray')
plt.subplot(2,2,2)
plt.imshow(v2[0].data,cmap = 'gray')
plt.subplot(2,2,3)
plt.imshow(v3[0].data,cmap = 'gray')
```

```python
# removing bias from every image using master bias
b1_ = b1[0].data - mb[0].data
b2_ = b2[0].data - mb[0].data
b3_ = b3[0].data - mb[0].data
v1_ = v1[0].data - mb[0].data
v2_ = v2[0].data - mb[0].data
v3_ = v3[0].data - mb[0].data
```

```python
# calibrating image by normalising to unity and taking average over total 
# for B
b1_norm = b1_/nfb[0].data
b2_norm = b2_/nfb[0].data
b3_norm = b3_/nfb[0].data
b_avg = (b1_norm + b2_norm + b3_norm)/100
# for V
v1_norm = v1_/nfv[0].data
v2_norm = v2_/nfv[0].data
v3_norm = v3_/nfv[0].data
v_avg = (v1_norm + v2_norm + v3_norm)/100
```

```python
#viewing the image again after calibration
plt.style.use('dark_background')
plt.imshow(b_avg,cmap = 'gray')
```

```python
# It seems the data is skewed.We take ln() to normalize
plt.imshow(np.log(b_avg),cmap = 'gray')
#Much better view 
```

```python
# It seems the data is skewed.We take ln()
# Similarly for V
plt.imshow(np.log(v_avg),cmap = 'gray')
#Much better view
```

```python
# Now we will design a function to find the location of the brightest stars in the given image.We will find the pixel location
# of the brighest points .We can define a number of brightest stars  for the same.
# This function takes in the calibrated image and processes it to give pixel location for given number of brightest stars.
def brightest_points(v_avg,no_points,error):
    v_avg_ = v_avg.copy()
    no_obtained = 0
    star_loc = []
    for i in range(0,no_points):
        star_loc.append(np.argwhere(v_avg_==v_avg_.max()))
        v_avg_[star_loc[i][0][0] - error:star_loc[i][0][0] + error,star_loc[i][0][1] - error:star_loc[i][0][1]+ error] = 0
    return np.array(star_loc).reshape(no_points,2)
```

```python
#Suppose we take 28 stars . NGC 6913 has 50 stars 
# henceforth
star_loc = brightest_points(v_avg,28,40)
```

```python
fig, ax = plt.subplots() 
plt.imshow(np.log(v_avg), cmap = 'gray' )
for loc in star_loc:
    x2,x1 = loc
    marker = plt.Circle((x1, x2), radius = 30, color='white',linewidth = 0.5 ,fill = False)
    ax.add_patch(marker)
```

```python
#identifying the stars (verifying) from image of b filter
fig, ax = plt.subplots() 
plt.imshow(np.log(b_avg), cmap = 'gray' )
for loc in star_loc:
    x2,x1 = loc
    marker = plt.Circle((x1, x2), radius = 30, color='pink',linewidth = 0.5 ,fill = False)
    ax.add_patch(marker)
```

```python
# A program to find out the B-V,V magnitudes 
def B_V_plotter(ref_magb,ref_magv,no_stars):
    M_b_mag = []
    for i in range(0,no_stars):
        temp_list = star_loc[i].tolist()
        M_b_mag.append(ref_magb - 2.5*np.log10(b_avg[temp_list[0],temp_list[1]]))
    M_v_mag = []
    for i in range(0,no_stars):
        temp_list2 = star_loc[i].tolist()
        M_v_mag.append(ref_magv - 2.5*np.log10(v_avg[temp_list2[0],temp_list2[1]]))
    b_v = np.array(M_b_mag) - np.array(M_v_mag)
    return b_v,M_v_mag,M_b_mag
```

```python
#the values have  been adjusted after trial and testing
b_v,M_v_mag,M_b_mag = B_V_plotter(11,12.5,28)
plt.scatter(b_v,M_v_mag)
plt.ylim(np.array(M_v_mag).max()+1,np.array(M_v_mag).min()-1)
plt.xlabel("B-V magnitude")
plt.ylabel("V Magnitude")
plt.title("Colour Magnitude Diagram")
```

With this, you know where the main sequence is supposed to be. Then, an H-R Diagram is made of the cluster or galaxy to which the distance is in question. Graphing the color and apparent magnitude, you get a similar H-R Diagram.</li>
